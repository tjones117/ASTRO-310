```python
# Import modules

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from astro310 import *

# (If you encounter problems with plots not appearing, then
# comment out the plot_defaults() command

plot_defaults()
```

```python
# Define physical constants

LSUN = 3.83E26
RSUN = 6.96E8
MSUN = 1.99E30

G = 6.67E-11
SIGMA = 5.67E-8
```

```python
# Load comma-separated variable (CSV) data from the 'corrected' tab of the
# adopt-a-star spreadsheet (https://tinyurl.com/yd4kkyot), saved as 'adopt.csv'
#
#  - this 'skip_header' parameter causes the first (header) row to be skipped
#  - the 'delimeter' parameter specifies how the columns are separated

data = np.genfromtxt('adopt.csv', skip_header=1, delimiter=',')

# This will tell us the dimensions of the data array (32 rows, 7 colums)

print(np.shape(data))
```

```python
# Have a look at the columns data

print(data[:,0])
print(data[:,1])
print(data[:,2])

# Columns 0 and 1 contain text; when python tries to interpret them as numbers, it can't, and so it
# fills them with the special value 'nan' (not a number)
```

```python
# Extract values from the columns (converting to SI units)

M = data[:,2]*MSUN
L = data[:,3]*LSUN
R = data[:,4]*RSUN
Teff = data[:,5]
t_half = data[:,6]*365.25*24*3600
```

```python
# Calculate and plot timescales

t_dyn = np.sqrt(R**3/(G*M))
t_thm = G*M**2/(R*L)
t_nuc = 2*t_half

plt.scatter(np.log10(M), np.log10(t_dyn), label='dyn')
plt.scatter(np.log10(M), np.log10(t_thm), label='thm')
plt.scatter(np.log10(M), np.log10(t_nuc), label='nuc')
plt.xlabel('log M (Msun)')
plt.ylabel('log timescale (s)')
plt.legend()
```
