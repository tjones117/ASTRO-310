```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from astro310 import *

plot_defaults()
```

```python
LSUN = 3.83E26
RSUN = 6.96E8
MSUN = 1.99E30
SIGMA = 5.67E-8
G = 6.67E-11
```

```python
#Load .csv file for the adopted stars

data = np.genfromtxt('adopt.csv', skip_header=1, delimiter=',')

print(np.shape(data))
```

```python
print(data[:, 0])
print(data[:, 1])
print(data[:, 2])
```

```python
M = data[:, 2]*MSUN
L = data[:, 3]*LSUN
R = data[:, 4]*RSUN
Teff = data[:, 5]
t_half = data[:, 6] * 365.25 * 24 * 3600
```

```python
t_dyn = np.sqrt(R**3/(G*M))
t_thm = G*M**2/(R*L)
t_nuc = 2*t_half

plt.scatter(np.log10(M), np.log10(t_dyn), label='dyn')
plt.scatter(np.log10(M), np.log10(t_thm), label='thm')
plt.scatter(np.log10(M), np.log10(t_nuc), label='nuc')
plt.xlabel('log M (Msun)')
plt.ylabel('log timescale (s)')
plt.legend()
```
